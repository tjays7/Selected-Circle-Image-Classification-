```python
import cv2
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
import numpy as np
```

```python
path = os.path.abspath('DS')
print(path)
```

```python
for img_file in os.listdir(path):  
            
        img = cv2.imread(os.path.join(path,img_file),0)
        cv2.imshow("image",img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        break  # because we want single image
```

```python
print(img.shape)
```

```python
IMG_SIZE_y = 22
IMG_SIZE_x = 90
for img_file in tqdm(os.listdir(path)):  
    img = cv2.imread(os.path.join(path,img_file),0)
    img_resize = cv2.resize(img, (IMG_SIZE_x, IMG_SIZE_y))
    cv2.imwrite(os.path.join(path,img_file),img_resize)
    
```

```python
i=0
for img_file in os.listdir(path):  
    img = cv2.imread(os.path.join(path,img_file),0)
    print(img.shape)
    i+=1
    if i>10:
        break
```

```python
labels = [0,1,2,3,4]
```

```python
training_data =[]

training_data = [[cv2.imread(os.path.join(path,img_file) ,0) ,labels[0]] if os.path.splitext(img_file)[0].endswith('a') 
                  else ([cv2.imread(os.path.join(path,img_file) ,0),labels[1]]) if os.path.splitext(img_file)[0].endswith('b')
                 else ([cv2.imread(os.path.join(path,img_file) ,0),labels[2]]) if os.path.splitext(img_file)[0].endswith('c')
                 else ([cv2.imread(os.path.join(path,img_file) ,0),labels[3]]) if os.path.splitext(img_file)[0].endswith('d')
                 else ([cv2.imread(os.path.join(path,img_file) ,0),labels[4]]) if os.path.splitext(img_file)[0].endswith('e')
                 else ''
                 for img_file in tqdm(os.listdir(path))]

print(training_data)
```
